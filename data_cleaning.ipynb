```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('glassdoor_jobs.csv')
df.head(30)
```

```python
# salary estimate
# company name
# job description
# state field
# age of company
```

```python
df["Salary Estimate"].value_counts()
```

```python
# Sal with -1 row are not needed 
df = df[df["Salary Estimate"] != '-1']
```

```python
salary = df["Salary Estimate"].apply(lambda x: x.split('(')[0].replace("$","").replace("K",""))
```

```python
df["hourly"] = df["Salary Estimate"].apply(lambda x: 1 if "per hour" in x.lower() else 0)
df["employer provided sal"] = df["Salary Estimate"].apply(lambda x: 1 if "employer provided salary:" in x.lower() else 0)
```

```python
df["hourly"].value_counts()
```

```python
df["employer provided sal"].value_counts()
```

```python
salary_hr = salary.apply(lambda x: x.lower().replace("per hour",""))
```

```python
salary_hr.apply(lambda x: len(x)>8 ).value_counts()
```

```python
salary_emp = salary_hr.apply(lambda x: x.lower().replace("employer provided salary:",""))
```

```python
salary_emp.apply(lambda x: len(x) >8).value_counts()
```

```python
df["min_sal"] = salary_emp.apply(lambda x: int(x.split("-")[0]))
```

```python
df["max_sal"] = salary_emp.apply(lambda x: int(x.split("-")[1]))
```

```python
df["avg_sal"] = (df.min_sal+df.max_sal)/2
```

```python
df
```

```python
# company name 

df['company_txt'] = df.apply(lambda x: x['Company Name'] if x["Rating"]<0 else x["Company Name"][:-4], axis = 1 )
```

```python
df
```

```python
#state field

df["job_state"] = df["Location"].apply(lambda x: x.split(',')[1])
```

```python
df["job_state"].value_counts()
```

```python
df["same_state"] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
```

```python
import time
import datetime
```

```python
# age of the company
df['age'] = df.Founded.apply(lambda x: x if x < 1 else time.localtime().tm_year - x)
```

```python
# job descriptions 
df['Job Description'][2].split("\n")
```

```python
df["python"] = df["Job Description"].apply(lambda x: 1 if "python" in x.lower() else 0 )
df["r_studio"] = df["Job Description"].apply(lambda x: 1 if "r studio" in x.lower() else 0 )
df["spark"] = df["Job Description"].apply(lambda x: 1 if "spark" in x.lower() else 0 )
df["aws"] = df["Job Description"].apply(lambda x: 1 if "aws" in x.lower() else 0 )
df["excel"] = df["Job Description"].apply(lambda x: 1 if "excel" in x.lower() else 0 )
```
